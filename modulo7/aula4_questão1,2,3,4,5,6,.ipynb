```python
def salvar_frase():
    frase = input("Digite uma frase: ")
    with open("frase.txt", "w") as arquivo:
        arquivo.write(frase)
    import os
    caminho_completo = os.path.abspath("frase.txt")
    print(f"Frase salva em {caminho_completo}")

salvar_frase()
```

```python
def processar_frase():
    with open("frase.txt", "r") as arquivo:
        conteudo = arquivo.read()
    palavras = ''.join([c if c.isalpha() or c.isspace() else ' ' for c in conteudo]).split()
    with open("palavras.txt", "w") as arquivo:
        for palavra in palavras:
            arquivo.write(palavra + "\n")
    with open("palavras.txt", "r") as arquivo:
        print(arquivo.read())

processar_frase()
```

```python
def analisar_roteiro():
    import requests

    url = "https://aplauso.imprensaoficial.com.br/edicoes/12.0.813.502/12.0.813.502.txt"
    response = requests.get(url)
    with open("estomago.txt", "w", encoding="utf-8") as arquivo:
        arquivo.write(response.text)

    with open("estomago.txt", "r", encoding="utf-8") as arquivo:
        linhas = arquivo.readlines()

    print("Primeiras 25 linhas:")
    for linha in linhas[:25]:
        print(linha.strip())

    print(f"Número total de linhas: {len(linhas)}")

    linha_maior = max(linhas, key=len)
    print(f"Linha com maior número de caracteres: {linha_maior.strip()}")

    nonato_count = sum(1 for linha in linhas if "nonato" in linha.lower())
    iria_count = sum(1 for linha in linhas if "íria" in linha.lower())
    print(f"Menções a 'Nonato': {nonato_count}")
    print(f"Menções a 'Íria': {iria_count}")

analisar_roteiro()
```

```python
import random

def carregar_palavras():
    with open("gabarito_forca.txt", "r") as arquivo:
        palavras = [linha.strip() for linha in arquivo.readlines()]
    return palavras

def carregar_enforcado():
    with open("gabarito_enforcado.txt", "r") as arquivo:
        estagios = [linha.strip() for linha in arquivo.readlines()]
    return estagios

def imprime_enforcado(erros, estagios):
    print(estagios[erros])

def jogar_forca():
    palavras = carregar_palavras()
    estagios = carregar_enforcado()
    palavra = random.choice(palavras)
    palavra_oculta = ["_" for _ in palavra]
    tentativas = 6
    erros = 0
    letras_tentadas = []

    while erros < tentativas and "_" in palavra_oculta:
        print("Palavra:", " ".join(palavra_oculta))
        letra = input("Digite uma letra: ").lower()

        if letra in letras_tentadas:
            print("Você já tentou essa letra.")
            continue

        letras_tentadas.append(letra)

        if letra in palavra:
            for i, char in enumerate(palavra):
                if char == letra:
                    palavra_oculta[i] = letra
        else:
            erros += 1
            imprime_enforcado(erros, estagios)
    
    if "_" not in palavra_oculta:
        print(f"Parabéns, você acertou! A palavra era: {palavra}")
    else:
        print(f"Você perdeu! A palavra era: {palavra}")

jogar_forca()
```

```python
import csv

def analisar_spotify():
    with open("spotify-2023.csv", "r", encoding="latin-1") as arquivo:
        leitor = csv.DictReader(arquivo)
        musicas = [linha for linha in leitor if 2012 <= int(linha['released_year']) <= 2022]

    mais_tocadas_por_ano = []
    for ano in range(2012, 2023):
        musicas_do_ano = [musica for musica in musicas if int(musica['released_year']) == ano]
        if musicas_do_ano:
            mais_tocada = max(musicas_do_ano, key=lambda x: int(x['streams']))
            mais_tocadas_por_ano.append([
                mais_tocada['track_name'],
                mais_tocada['artist(s)_name'],
                mais_tocada['released_year'],
                int(mais_tocada['streams'])
            ])

    print(mais_tocadas_por_ano)

analisar_spotify()
```
