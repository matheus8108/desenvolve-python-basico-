```python
#Questão 1
import random

lista1=[]

for i in range(20):
    valor=random.randint(-100,100)
    lista1.append(valor)

print(sorted(lista1))
print(lista1)
print("O maior valor no indice é:",lista1.index(max(lista1)))
print("O menor valor no indice é:",lista1.index(min(lista1)))
```

```python
#Questão 3
import random

lista1=[]
lista2=[]
interseccao=[]

for i in range(20):
    valor1=random.randint(-100,100)
    lista1.append(valor1)
    valor2=random.randint(-100,100)
    lista2.append(valor2)


for elemento in lista1:  
    for elemento in lista1:
        interseccao.append(elemento)  

print(lista1)
print(lista2)
print(interseccao)


interseccao.sort()
for i in interseccao:
    print(f"{i}:({lista1.count(i)},{lista2.count(i)})")
```

```python
#Questão 4

ista1 = []
lista2 = []

tamanho1 = int(input("Quantos números na primeira lista? "))
for i in range(tamanho1):
    lista1.append(int(input("Digite um número: ")))

tamanho2 = int(input("Quantos números na segunda lista? "))
for i in range(tamanho2):
    lista2.append(int(input("Digite um número: ")))

lista_intercalada = []
for num1, num2 in zip(lista1, lista2):
    lista_intercalada.append(num1)
    lista_intercalada.append(num2)

lista_intercalada.extend(lista1[len(lista2):])
lista_intercalada.extend(lista2[len(lista1):])

print("Lista intercalada:", ' '.join(map(str, lista_intercalada)))
```
